```python
import os
import sqlite3
from faker import Faker
import random

# Nombre del archivo de la base de datos
db_name = "bulk.db"

# Verificar si el archivo de la base de datos existe
if os.path.exists(db_name):
    # Cerrar cualquier conexión abierta (si existe)
    conn = sqlite3.connect(db_name)
    conn.close()

    # Eliminar el archivo de la base de datos
    os.remove(db_name)
    print(f"La base de datos '{db_name}' ha sido eliminada.")
else:
    print(f"La base de datos '{db_name}' no existe.")


# Crear conexión SQLite y tabla
conn = sqlite3.connect(db_name)
cursor = conn.cursor()

# Crear tabla
cursor.execute("""
CREATE TABLE IF NOT EXISTS large_table (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    email TEXT,
    address TEXT,
    country TEXT,
    gender TEXT,
    age INTEGER,
    profession TEXT,
    zodiac_sign TEXT,
    favorite_food TEXT,
    favorite_sport TEXT,
    favorite_movie_genre TEXT,
    favorite_animal TEXT,
    preferred_language TEXT,
    hobby TEXT,
    favorite_tv_show TEXT,
    favorite_color TEXT
)
""")
conn.commit()

# Generar datos falsos con Faker
fake = Faker()
records_to_insert = 5_000_000  # Número total de registros
batch_size = 50_000  # Insertar registros en lotes para mejor rendimiento
numero_de_lote = 1
for batch_start in range(0, records_to_insert, batch_size):
    batch = [
        (
            fake.email(),
            fake.address(),
            fake.country(),
            random.choice(genders),
            random.randint(18, 99),
            random.choice(professions),
            random.choice(zodiac_signs),
            random.choice(favorite_foods),
            random.choice(favorite_sports),
            random.choice(movie_genres),
            random.choice(animals),
            random.choice(languages),
            random.choice(hobbies),
            random.choice(tv_shows),
            random.choice(colors),
        )
        for _ in range(batch_size)
    ]
    cursor.executemany(
        """
        INSERT INTO large_table (
            email, address, country, gender, age, profession, zodiac_sign, favorite_food, 
            favorite_sport, favorite_movie_genre, favorite_animal, preferred_language,
            hobby, favorite_tv_show, favorite_color
        ) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
        """,
        batch
    )
    conn.commit()
    print(f"Lote insertado {numero_de_lote} : {batch_start + batch_size} registros han sido guardados")
    numero_de_lote +=1

conn.close()
print(f"Base de datos SQLite creada con {records_to_insert} registros en '{db_name}'.")
```
