```python
import random
from faker import Faker

fake = Faker()

def generate_fake_data(records):
    for _ in range(records):
        yield {
            "email": fake.email(),
            "address": fake.address(),
            "country": fake.country(),
            "sex": random.choice(sexes),
            "age": random.randint(18, 99),
            "profession": random.choice(professions),
            "zodiac_sign": random.choice(zodiac_signs),
            "favorite_food": random.choice(favorite_foods),
            "favorite_sport": random.choice(favorite_sports),
            "favorite_movie_genre": random.choice(movie_genres),
            "favorite_animal": random.choice(animals),
            "preferred_language": random.choice(languages),
            "hobby": random.choice(hobbies),
            "favorite_tv_show": random.choice(tv_shows),
            "favorite_color": random.choice(colors),
            "favorite_drink": random.choice(drinks),
            "favorite_music": random.choice(music_genres),
            "favorite_technology": random.choice(technologies),
            "favorite_car": random.choice(cars),
        }
```

# FUNCIONES

```python
import sqlite3
import os
import json
import os
import math
import pandas as pd

def save_to_sqlite_with_generator(data_generator, total_records, db_name, batch_size=10_000, directory="/sqlite_database"):
    """
    Guarda los datos generados en una base de datos SQLite en lotes, usando un generador para manejar grandes vol√∫menes de datos.

    Args:
        data_generator (generator): Generador que produce los datos.
        total_records (int): N√∫mero total de registros a generar.
        db_name (str): Nombre de la base de datos SQLite.
        batch_size (int): N√∫mero de registros por lote.
        directory (str): Directorio donde guardar la base de datos.
    """
    # Crear el directorio si no existe
    os.makedirs(directory, exist_ok=True)
    db_path = os.path.join(directory, db_name)

    # Verificar si el archivo de la base de datos existe y eliminarlo si es necesario
    if os.path.exists(db_path):
        os.remove(db_path)
        print(f"La base de datos '{db_path}' ha sido eliminada.")
    
    # Crear conexi√≥n SQLite y tabla
    try:
        conn = sqlite3.connect(db_path)
    except sqlite3.OperationalError as e:
        print(f"Error al conectar con la base de datos: {e}")
        return

    cursor = conn.cursor()
    cursor.execute("""
    CREATE TABLE IF NOT EXISTS large_table (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        email TEXT,
        address TEXT,
        country TEXT,
        sex TEXT,
        age INTEGER,
        profession TEXT,
        zodiac_sign TEXT,
        favorite_food TEXT,
        favorite_sport TEXT,
        favorite_movie_genre TEXT,
        favorite_animal TEXT,
        preferred_language TEXT,
        hobby TEXT,
        favorite_tv_show TEXT,
        favorite_color TEXT,
        favorite_drink TEXT,
        favorite_music TEXT,
        favorite_technology TEXT,
        favorite_car TEXT
    )
    """)
    conn.commit()

    # Insertar datos en lotes
    batch = []
    for i, record in enumerate(data_generator, start=1):
        batch.append(record)
        if len(batch) == batch_size or i == total_records:
            cursor.executemany(
                """
                INSERT INTO large_table (
                    email, address, country, sex, age, profession, zodiac_sign, favorite_food, 
                    favorite_sport, favorite_movie_genre, favorite_animal, preferred_language,
                    hobby, favorite_tv_show, favorite_color, favorite_drink, favorite_music, 
                    favorite_technology, favorite_car
                ) VALUES (
                    :email, :address, :country, :sex, :age, :profession, :zodiac_sign, 
                    :favorite_food, :favorite_sport, :favorite_movie_genre, :favorite_animal, 
                    :preferred_language, :hobby, :favorite_tv_show, :favorite_color,
                    :favorite_drink, :favorite_music, :favorite_technology, :favorite_car
                )
                """,
                batch
            )
            conn.commit()
            print(f"Batch {i // batch_size + 1} insertado.")
            batch = []

    conn.close()
    print(f"Datos guardados en la base de datos '{db_path}'.")

def save_to_csv_with_generator(data_generator, total_records, num_files, csv_file_prefix="bulk_data", directory="./csv_files"):
    os.makedirs(directory, exist_ok=True)
    batch_size = math.ceil(total_records / num_files)

    for file_index in range(num_files):
        file_path = os.path.join(directory, f"{csv_file_prefix}_{file_index + 1}.csv")
        batch = [next(data_generator) for _ in range(min(batch_size, total_records - file_index * batch_size))]
        df = pd.DataFrame(batch)
        df.to_csv(file_path, sep=";", index=False)
        print(f"Datos guardados en el archivo CSV '{file_path}' > {file_index + 1} de {num_files}.")


def save_to_json_with_generator(data_generator, total_records, num_files, json_file_prefix="bulk_data", directory="./json_files"):
    os.makedirs(directory, exist_ok=True)
    batch_size = math.ceil(total_records / num_files)

    for file_index in range(num_files):
        file_path = os.path.join(directory, f"{json_file_prefix}_{file_index + 1}.json")
        batch = [next(data_generator) for _ in range(min(batch_size, total_records - file_index * batch_size))]
        with open(file_path, "w", encoding="utf-8") as json_file:
            json.dump(batch, json_file, indent=4)
        print(f"Datos guardados en el archivo JSON '{file_path}' > {file_index + 1} de {num_files}.")

def save_to_parquet_with_generator(data_generator, total_records, num_files, parquet_file_prefix="bulk_data", directory="./parquet_files"):
    os.makedirs(directory, exist_ok=True)
    batch_size = math.ceil(total_records / num_files)

    for file_index in range(num_files):
        file_path = os.path.join(directory, f"{parquet_file_prefix}_{file_index + 1}.parquet")
        batch = [next(data_generator) for _ in range(min(batch_size, total_records - file_index * batch_size))]
        df = pd.DataFrame(batch)
        df.to_parquet(file_path, index=False)
        print(f"Datos guardados en el archivo Parquet '{file_path}' > {file_index + 1} de {num_files}.") 
```
